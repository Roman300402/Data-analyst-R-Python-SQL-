###
* Usando agg, calcule cual era la población total de todo el país en el 2020, recuerde que la población de toda la localidad esta en la columna POBTOT.

```python
poblacion_total_2020 = data.agg({'POBTOT': 'sum'})['POBTOT']
```

```python

print(f"La poblacion total en el año 2020 es {poblacion_total_2020}" )
```

###
* Compare la población total masculina y la población total femenina de todo el país e indique si la "probabilidad" real de que un bebe sea hombre o mujer si es 50-50 o se acerca a ese equilibrio, usando la columna POBFEM y POBMAS.

> ⚠️ **Advertencia:**  Debido a que en algunas celdas de la base de datos hay * no se puede realizar la suma de manera inmediata por tal razón se realiza el siguiente proceso con replace donde cambiaremos a * por ceros y luego lo volvemos a tipo numerico 


```python
data['POBFEM'] = data['POBFEM'].replace('*', 0) 
data['POBMAS'] = data['POBMAS'].replace('*',0)
data['POBFEM'] = pd.to_numeric(data['POBFEM'], errors='coerce').fillna(0)
data['POBMAS'] = pd.to_numeric(data['POBMAS'], errors='coerce').fillna(0)

poblacion_total_femenina = data['POBFEM'].sum()
poblacion_total_masculina = data['POBMAS'].sum()
```

> **Nota:** Por lo anteriormente explicado entonces debemos volver a calcular la población total sumando tanto la poblacion femenina como masculina , esto es permitenente ya que si queremos obtener la probabilidad de que si nace un bebe sea femenino o masculino su espacio muestral total tiene que ser igual a sumar ambos.


```python
poblacion_total = poblacion_total_femenina + poblacion_total_masculina
probabilidad_femenina = poblacion_total_femenina / poblacion_total
probabilidad_masculina = poblacion_total_masculina / poblacion_total
```

```python
print(f"Población total femenina: {poblacion_total_femenina}")
print(f"Población total masculina: {poblacion_total_masculina}")
print(f"Probabilidad de que un bebé sea mujer: {probabilidad_femenina:.2%}")
print(f"Probabilidad de que un bebé sea hombre: {probabilidad_masculina:.2%}")
```

###
* Indique cual estado es el que tiene mas habitantes totales. (Tip: Use groupby para sumar la población por entidad y luego indique cual es la que mas pobladores tiene)

```python

poblacion_por_estado = data.groupby('NOM_ENT')['POBTOT'].sum().reset_index()
estado_con_mas_habitantes = poblacion_por_estado.loc[poblacion_por_estado['POBTOT'].idxmax()]

```

```python

print(f"El estado con más habitantes es {estado_con_mas_habitantes['NOM_ENT']} con una población de {estado_con_mas_habitantes['POBTOT']}")
```

###
* La columna VIVTOT indica cuantas viviendas existen en esa localidad, indique cual estado es el que tiene mas viviendas en el (2020).

```python
vivienda_por_estado = data.groupby('NOM_ENT')['VIVTOT'].sum().reset_index()
estado_con_mas_viviendas = vivienda_por_estado.loc[vivienda_por_estado['VIVTOT'].idxmax()]
```

```python
print(f"El estado con más viviendas es {estado_con_mas_viviendas['NOM_ENT']} con una población de {estado_con_mas_viviendas['VIVTOT']}")
```

###
* Cuantas viviendas hay en la Ciudad de México? 

```python

ciudad_mexico_data = data[data['NOM_ENT'] == 'Ciudad de México']
vivienda_ciudad_mexico = ciudad_mexico_data.groupby('NOM_ENT')['VIVTOT'].sum().reset_index()
```

```python
print(f"Las viviendas totales de la Ciudad de México es de {vivienda_ciudad_mexico['VIVTOT'].values[0]}")
```

###
* Cuantos habitantes hay en la Ciudad de México?

```python
ciudad_mexico_data = data[data['NOM_ENT'] == 'Ciudad de México']
poblacion_ciudad_mexico = ciudad_mexico_data.groupby('NOM_ENT')['POBTOT'].sum().reset_index()
```

```python
print(f"La población total de la Ciudad de México es {poblacion_ciudad_mexico['POBTOT'].values[0]}")
```

###
* Cuantas viviendas por persona hay en la Ciudad de México? Es decir de la razón de viviendas vs habitantes.

```python
vivienda_por_persona= vivienda_ciudad_mexico['VIVTOT'].values[0] / poblacion_ciudad_mexico['POBTOT'].values[0]
```

```python
print(f"En la Ciudad de México hay un proporción de {vivienda_por_persona} viviendas por persona")
```

###
* (Opcional) Cual estado de México tiene mas viviendas por persona, es decir siguiendo la ley de oferta-demanda, en teoría en ese estado debería ser mas barato comprar casa... en teoría.

```python

poblacion_y_viviendas_por_estado = data.groupby('NOM_ENT')[['POBTOT', 'VIVTOT']].sum().reset_index()

poblacion_y_viviendas_por_estado['VIV_POR_PERSONA'] = poblacion_y_viviendas_por_estado['VIVTOT'] / poblacion_y_viviendas_por_estado['POBTOT']

estado_con_mas_viv_por_persona = poblacion_y_viviendas_por_estado.loc[poblacion_y_viviendas_por_estado['VIV_POR_PERSONA'].idxmax()]

```

```python
print(f"El estado con más viviendas por persona es {estado_con_mas_viv_por_persona['NOM_ENT']} con una proporción de {estado_con_mas_viv_por_persona['VIV_POR_PERSONA']:.2f} viviendas por persona.")
```
